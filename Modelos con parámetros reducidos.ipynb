```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from itertools import combinations
```

# Bosque Aleatorio por Regresión

```python
df_combi=pd.read_excel('INSF.xlsx')
columnas=df_combi.columns[2:11]

X=df_combi[columnas]
y=df_combi['INSF']

combi_columnas=[c for i in range(1, len(columnas) + 1) for c in combinations(columnas, i)]

resultados={}

for combo in combi_columnas:
    X_combi=X[list(combo)]

    modelo_combinado=RandomForestRegressor()
    valcru_combi=cross_val_score(modelo_combinado, X_combi, y, cv=10)
    resultados[combo]=valcru_combi
    print(f'Columnas: {combo}')
    print('Validación cruzada del Bosque Aleatorio Combinado:', valcru_combi)
    print('Promedio:', valcru_combi.mean(), '-', 'Desviación estándar:', valcru_combi.std())
    print('-'*50)
```

```python
df=pd.DataFrame.from_dict(resultados,orient='index')
df.to_excel('Resultados.xlsx')
```
