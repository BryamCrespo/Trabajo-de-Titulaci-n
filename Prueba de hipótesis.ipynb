```python
import pandas as pd
from scipy.stats import ttest_ind
```

## Prueba T con valores mayores a 0.94 (Valores que sobrepasan al conjunto original)

```python
datos=pd.read_excel('Resultados 0.94.xlsx', index_col=0)

conjunto_original=datos.iloc[-1]
otros_conjuntos=datos.iloc[:-1]

resultados=[ttest_ind(conjunto_original, otro_conjunto) for _, otro_conjunto in otros_conjuntos.iterrows()]

for i, (conjunto_nombre, resultado_t) in enumerate(zip(otros_conjuntos.index, resultados), start=1):
    estadistico_t=resultado_t.statistic
    valor_p=resultado_t.pvalue

    print(f'Prueba t para {conjunto_nombre}:')
    print(f'Estadístico t: {estadistico_t}')
    print(f'Valor p: {valor_p}')
    print('Resultado: Diferentes conjuntos' if valor_p < 0.05 else 'No hay evidencia para rechazar la hipótesis nula')
    print('='*50)
```

## Prueba T con valores mayores a 0.90

```python
datos=pd.read_excel('Resultados 0.90.xlsx', index_col=0)

conjunto_original=datos.iloc[-1]
otros_conjuntos=datos.iloc[:-1]

resultados=[ttest_ind(conjunto_original, otro_conjunto,equal_var=False) for _, otro_conjunto in otros_conjuntos.iterrows()]

for i, (conjunto_nombre, resultado_t) in enumerate(zip(otros_conjuntos.index, resultados), start=1):
    estadistico_t=resultado_t.statistic
    valor_p=resultado_t.pvalue

    print(f'Prueba t para {conjunto_nombre}:')
    print(f'Estadístico t: {estadistico_t}')
    print(f'Valor p: {valor_p}')
    print('Resultado: Diferentes conjuntos' if valor_p < 0.05 else 'No hay evidencia para rechazar la hipótesis nula')
    print('='*50)
```
